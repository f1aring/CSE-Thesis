```python
out = ds['Log Price'] # label as target # y
input = ds.drop(['Log Price'], axis = 1) # x
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(input,out)
```

```python
plt.figure(figsize=[12,6])
feat_importances = pd.Series(model.feature_importances_, index= input.columns)
feat_importances.nlargest(6).plot(kind='barh')
plt.show()
```

```python
sc = StandardScaler()
```

```python
sc.fit(input)
```

```python
feature = sc.transform(input) 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(feature, out, test_size=0.20, random_state=140)
print("X_train shape = ", X_train.shape)
print("X_test shape = ", X_test.shape)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train,Y_train)
RF_accuracy = rf.score(X_test, Y_test)
print('Random Forrest accuracy: {:.2f}'.format(RF_accuracy*100))
```

```python
RF_prediction = rf.predict(X_test)
RF_accuracy =r2_score(Y_test,RF_prediction) 
print(RF_accuracy)
```

```python
print('MAE:', metrics.mean_absolute_error(Y_test, RF_prediction))
print('MSE:', metrics.mean_squared_error(Y_test, RF_prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, RF_prediction)))
```

```python
nn = MLPClassifier()
Y_train_int=Y_train.astype(int)
Y_test_int=Y_test.astype(int)
nn.fit(X_train,Y_train_int)
NNaccuracy = nn.score(X_test, Y_test_int)
print('Neural network accuracy: {:.2f}'.format(NNaccuracy*100))
```

```python
NN_prediction = nn.predict(X_test)
NN_accuracy = r2_score(Y_test_int,NN_prediction) 
print(NN_accuracy)
```

```python
print('MAE:', metrics.mean_absolute_error(Y_test_int, NN_prediction))
print('MSE:', metrics.mean_squared_error(Y_test_int, NN_prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test_int, NN_prediction)))
```

```python
svc_model = SVC(kernel="linear")
Y_train_int=Y_train.astype(int)
Y_test_int=Y_test.astype(int)
svc_model.fit(X_train,Y_train_int)
svc_model_accuracy = svc_model.score(X_test, Y_test_int)
print('SVM accuracy: {:.2f}'.format(svc_model_accuracy*100))
```

```python
SVM_prediction = svc_model.predict(X_test)
SVM_accuracy = r2_score(Y_test_int,SVM_prediction) 
print(SVM_accuracy)
```

```python
print('MAE:', metrics.mean_absolute_error(Y_test_int, SVM_prediction))
print('MSE:', metrics.mean_squared_error(Y_test_int, SVM_prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test_int, SVM_prediction)))
```

```python
y_ticks = [x / 100 for x in range(14, 108, 12)]
plt.bar(
    height=[RF_accuracy, NNaccuracy, svc_model_accuracy],
    x=['Random forest', 'Neuralo network (MLP)', 'Support Vector Machine'],
    color=['Purple', 'grey', 'Red'])
plt.yticks(y_ticks, fontsize=8)
plt.ylabel("Accuracy levels")
colors = {'Random forest': 'Purple', 'Neuralo network (MLP)': 'grey','Support Vector Machine': 'Red'}
labels = list(colors.keys())
handles = [plt.Rectangle((0, 0), 1, 1, color=colors[label]) for label in labels]
plt.legend(handles, labels)
plt.title("Accuracy chart (Random forest vs Neuralo network (MLP) vs Support Vector Machine)", fontsize=15)
plt.show()
```
